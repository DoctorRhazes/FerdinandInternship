```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

HTMLCode = urlopen("https://www.pythonscraping.com/pages/page3.html")
BSObjekt = BeautifulSoup(HTMLCode,"lxml") # Code direkt in BS Objekt umgewandelt

for kind in BSObjekt.find("table",{"id":"giftList"}).children:
    print(kind)
# *.children sind die DIREKTEN Unterpunkte vom Objekt .find()
# find() findet: Objekt "table" wo "id=giftList"
# .children ist ein list_iterator Objekt -> kann nicht als Liste mit [] verwendet werden!
# kind = BSObjekt.find("table",{"id":"giftList"}).children[0] -> ERROR
```

```python
for Nachkomme in BSObjekt.find("table",{"id":"giftList"}).tr.next_siblings:
    print(Nachkomme)
# descendants: children inkl. children von children in endloser rekursiver Reihen, dh Unterpunkte auf allen Levels
```

```python
tdTags = BSObjekt.findAll(lambda tr: len(tr.attrs)==2)
print(tdTags[0])
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

html = urlopen("http://en.wikipedia.org/wiki/Kevin_Bacon")
bsObj = BeautifulSoup(html,"lxml")

for link in bsObj.find("div", {"id":"bodyContent"}).findAll("a",
href=re.compile("^(/wiki/)((?!:).)*$")):
    if 'href' in link.attrs:
        print(link.attrs['href'])
        
```
